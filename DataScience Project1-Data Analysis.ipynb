## Duplicate data removal 

```python
duplicate_apps=[]
unique_apps=[]
count=0

for app in google_apps_data:
    name=app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
        count+=1
        
    else:
        unique_apps.append(name)
if(count==len(duplicate_apps)):
    print('Count=',count)
```

```python
for app in google_apps_data:
    name=app[0]
    if name in duplicate_apps:
        print(app[:5])
      
```

```python
reviews_max={}
for app in google_apps_data:
    name=app[0]
    n_reviews=float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name]=n_reviews
    elif name not in reviews_max:
        reviews_max[name]=n_reviews
        
```

```python
print(len(reviews_max))
```

```python
android_clean=[]
already_added=[]
for app in google_apps_data:
    name=app[0]
    n_reviews=float(app[3])
    if n_reviews == reviews_max[name] and name not in already_added:
        android_clean.append(app)
        already_added.append(name)
        
```

```python
def non_english(string):
    for char in string:
        if ord(char) > 127:
            return False
    return True   
```

## to find the the ascii value of character in string we use ord() and check it's value in range(0,127) which is the range of English Characters

```python
print(non_english("Instagram"))
print(non_english("爱奇艺PPS -《欢乐颂2》电视剧热播"))
```

```python
print(non_english("Docs To Go™ Free Office Suite"))
print(non_english("Instachat 😜"))
```

##  we get false as the ord(™) and ord(😜) is greater than 127 but Instachat is an English word so we need to ignore the special characters like this.

```python
apps_english=[]
apps_data_english=[]

for app in android_clean:
    name=app[0]
    if non_english(name):
        apps_english.append(app)
        
for app1 in apps_data:
    name1=app1[0]
    if non_english(name1):
        apps_data_english.append(app1)
        
        
```
